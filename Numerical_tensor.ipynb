```python
class opt:
    def __init__(self,dim,base,cut):
        self.dim=dim
        self.cut=cut
        self.base=base[:self.cut]
    
   
def othonormal_basis(level):
    baseset={key:[] for key in opset}
    p=cm.descendants(level)
    for op,dim in opset.items():
        for l in range(level+1):
            baseset[op].extend(p.othbasis(l,dim))
    return baseset
        
def tencal(o1,o2,o3):        # function to compute the rank-3 tensor 
    ten=np.zeros([o1.cut,o2.cut,o3.cut])
    for i,base1 in enumerate(o1.base):
        for j,base2 in enumerate(o2.base):
            for k,base3 in enumerate(o3.base):
                op1=cm.operator(dim=o1.dim,level=base1,xyz=0)
                op1.trans(h1)
                op2=cm.operator(dim=o2.dim,level=base2,xyz=0)
                op2.trans(h2)
                op3=cm.operator(dim=o3.dim,level=base3,xyz=0)
                op3.trans(h3)
                t=cm.tensor(op1,op2,op3)
                tva=N(t.value())
                ten[i][j][k]=tva
                print((i,j,k))
                print(tva)
    return ten

    
```

# Compute rank-3 tensor 

```python
# three conformal transformations (Taylor expansion) to compute the rank3 tensor 

colist=([3.23867657771728,
  2.52956375344893,
  1.51174590003830,
  1.47959741660741,
  1.21691543886601,
  1.19819933351408,
  1.06386832469529,
  1.05082937994331,
  0.965380442255512,
  0.955454668256735],
 [0.308768095857485,
  0.241162883900265,
  0.0442334958055958,
  0.0630402322562133,
  0.0227481182359890,
  0.0321676636387716,
  0.0144809557538896,
  0.0203299617293628,
  0.0102713487315804,
  0.0143271704150594],
 [-1.00000000000000,
  1.33333333333333,
  -0.888888888888889,
  0.839506172839506,
  -0.724279835390946,
  0.696844993141290,
  -0.637707666514252,
  0.619214550627444,
  -0.581483174990266,
  0.567756972708793])

h1=Lambda(x,sum([colist[0][n]*x**n for n in range(10)]))
h2=Lambda(x,sum([colist[1][n]*x**n for n in range(10)]))
h3=Lambda(x,sum([colist[2][n]*x**n for n in range(10)]))
```

```python
# compute the rank-3 tensor of three transformed descendants 

op1=cm.operator(dim=0,level=baseset['I'][0],xyz=0)
op1.trans(h1)
op2=cm.operator(dim=0,level=baseset['I'][1],xyz=0)
op2.trans(h2)
op3=cm.operator(dim=0,level=baseset['I'][2],xyz=0)
op3.trans(h3)
cm.tensor(op1,op2,op3).value()
```

```python
T = {}

cut_dic = {'I':2, 'E':3, 'S':3}     # states up to level 2


for triple in opeset:
    if not triple in T:
        print('now computing '+ triple)
        opt_lis = [opt(opset[op],baseset[op],cut_dic[op]) for op in triple]
        T[triple] = tencal(*opt_lis)
```
